```python
from Noise import Noise
import torch
import matplotlib.pyplot as plt

obj = Noise(300)
```

```python
x = torch.linspace(0,10,256)
signal = torch.sin(x)
plt.plot(x, signal)
```

```python
for idx in range(0, 300, 30):
    t = torch.Tensor([idx]).type(torch.int64)
    sig, noise = obj.forward_diffusion_sample(signal, t)
    plt.figure()
    plt.plot(x, sig.cpu())
    plt.show()
```
